```python
import sqlite3
from sqlite3 import Error
import pandas as pd
```

```python
#Q10
conn = sqlite3.connect('sqlite_db_pythonsqlite.db')
cur = conn.cursor()
q = """
        SELECT name as Facility_Name, SUM(cost) AS Revenue FROM 
            (SELECT Facilities.facid, Facilities.name,
            CASE
                WHEN Bookings.memid = CAST(0 as integer)
                    THEN CAST(Facilities.guestcost as integer) * CAST(slots as integer)
                ELSE CAST(Facilities.membercost as integer) * CAST(slots as integer)
            END as 'cost'
            FROM Bookings
            INNER JOIN Facilities ON Bookings.facid = Facilities.facid)
        GROUP BY facid
        HAVING Revenue < 1000
        ORDER BY Revenue DESC
    """
pd.read_sql(q, conn)
```

```python
#Q12 Find the facilities with their usage by member, but not guests
conn = sqlite3.connect('sqlite_db_pythonsqlite.db')
cur = conn.cursor()
q = """
    SELECT name as Facility_Name, memid
        From 'Bookings'
    INNER JOIN Facilities
    ON Bookings.facid = Facilities.facid
    """
pd.read_sql(q, conn)
#join members and bookings, then search in facilities 
```

```python
#Q13: Find the facilities usage by month, but not guests
conn = sqlite3.connect('sqlite_db_pythonsqlite.db')
cur = conn.cursor()
q = """
    SELECT name as Facility_Name, starttime
        From 'Bookings'
    INNER JOIN Facilities
    ON Bookings.facid = Facilities.facid
    ORDER BY starttime
    """
pd.read_sql(q, conn)
```
